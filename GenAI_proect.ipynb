```python
essay = """
The Indian Budget is an annual financial statement presented by the Government of India.
It outlines the government's revenue and expenditure for the upcoming fiscal year.
The budget is a crucial document that reflects the economic policies and priorities of the government.
It includes various components such as tax proposals, allocation of funds to different sectors, and measures to boost economic growth.
The budget also addresses issues like inflation, unemployment, and social welfare. Overall,
the Indian Budget plays a significant role in shaping the country's economic landscape.
"""
```

```python

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

#nltk.download('punkt')
nltk.download('stopwords')

def clean_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Convert to lowercase
    text = text.lower()
    print(text)
    # Tokenize text
    tokens = word_tokenize(text)
    print(tokens)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    print(stop_words)
    cleaned_tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(cleaned_tokens)

cleaned_essay = clean_text(essay)
print(cleaned_essay)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Tokenize and create embeddings
inputs = tokenizer(cleaned_essay, return_tensors='pt')
outputs = model(**inputs)
embeddings = outputs.last_hidden_state
print(embeddings)
```

```python
def generate_response(prompt):
    response = openai.chatCompletion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=100,
        n=1,
        stop=none,
        temperature=0.4,
    )
    return response.choices[0].text.strip()
```

```python
def genai_project(essay, question):
    cleaned_essay = clean_text(essay)
    inputs = tokenizer(cleaned_essay, return_tensors='pt')
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state
    response = generate_response(question)
    return response

##################################################################################

# Example usage
question = "What are the components of the Indian Budget?"
response = genai_project(essay, question)
print(response)
```
